```python
import paho.mqtt.client as mqtt

import time
import random

def on_connect(client, userdata, flags, rc):
    print("Connected with result code "+str(rc))

def on_publish(client, userdata, mid):
    print("mid: "+str(mid))
 
client2 = mqtt.Client()
client2.on_publish = on_publish
client2.on_connect = on_connect
#client2.connect('broker.mqttdashboard.com', 1883)
client2.username_pw_set('ua1','ua1')
client2.connect('128.196.142.13', 1883)

client2.loop_start()

while True:
    temperature = random.uniform(15, 35)
    (rc, mid) = client2.publish('temperature/sensor1', str(temperature), qos=1)
    (rc, mid) = client2.publish('temperature/sensor2', str(temperature+3), qos=1)
    time.sleep(4)
```
